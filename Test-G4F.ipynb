```python
import g4f

# メッセージのリストを作成
messages = [
    {"role": "user", "content": "こんにちは"}
]

# モデルを指定してチャットコンプリーションを作成
response = g4f.ChatCompletion.create(model='gpt-3.5-turbo', messages=messages)

# レスポンスを出力
for message in response:
    print(message['content'])
```
